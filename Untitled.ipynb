```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
with open('movie_data.txt') as json_file:
    data = json.load(json_file)

data['1999']['The Sixth Sense']
```

```python
imdb_rating = {}
imdb_list = []

for year in list(data.keys()):
    movies = list(data[year].keys())
    imdb_list = []
    for movie in movies:
        try:
            imdb_r = data[year][movie]['imdbRating']
            
            imdb_list.append(imdb_r)
        except:
            print(" Year: "+year+" Movie: "+movie)
    imdb_rating[year] = imdb_list
imdb_rating['1980']
```

```python
box_office = {}
box_list = []

for year in list(data.keys()):
    movies = list(data[year].keys())
    box_list = []
    for movie in movies:
        try:
            box_r = data[year][movie]['BoxOffice']
            
            box_list.append(box_r)
        except:
            print(" Year: "+year+" Movie: "+movie)
    box_office[year] = box_list
box_office['1980']
```
